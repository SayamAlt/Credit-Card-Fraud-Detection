```python
from flask import Flask, request, render_template
import joblib 
```

```python
app = Flask(__name__)
model = joblib.load('model.pkl')
```

```python
@app.route("/")
def home():
    return render_template("home.html")
```

```python
@app.route("/predict", methods=["GET","POST"])
def predict():
    if request.method == "POST":
        credit_card_no = request.form["ccn"]
        trans_amt = request.form["ta"]
        latitude = request.form["lat"]
        longitude = request.form["long"]
        city_pop = request.form["city_pop"]
        merch_lat = request.form["merch_lat"]
        merch_long = request.form["merch_long"]
        trans_reason = request.form["steps"]
        
        food_dining = 0
        gas_transport = 0
        grocery_net = 0
        grocery_pos = 0
        health_fitness = 0
        home = 0
        kids_pets = 0
        misc_net = 0
        misc_pos = 0
        personal_care = 0
        shopping_net = 0
        shopping_pos = 0
        travel = 0
        gender = 0
        
        if(trans_reason == "Food & Dining"):
            food_dining = 1
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 0
            home = 0
            kids_pets = 0
            misc_net = 0
            misc_pos = 0
            personal_care = 0
            shopping_net = 0
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Gas Transport"):
            food_dining = 0
            gas_transport = 1
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 0
            home = 0
            kids_pets = 0
            misc_net = 0
            misc_pos = 0
            personal_care = 0
            shopping_net = 0
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Grocery Net"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 1
            grocery_pos = 0
            health_fitness = 0
            home = 0
            kids_pets = 0
            misc_net = 0
            misc_pos = 0
            personal_care = 0
            shopping_net = 0
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Grocery Point of Sale(POS)"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 1
            health_fitness = 0
            home = 0
            kids_pets = 0
            misc_net = 0
            misc_pos = 0
            personal_care = 0
            shopping_net = 0
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Health & Fitness"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 1
            home = 0
            kids_pets = 0
            misc_net = 0
            misc_pos = 0
            personal_care = 0
            shopping_net = 0
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Home"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 0
            home = 1
            kids_pets = 0
            misc_net = 0
            misc_pos = 0
            personal_care = 0
            shopping_net = 0
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Kids & Pets"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 0
            home = 0
            kids_pets = 1
            misc_net = 0
            misc_pos = 0
            personal_care = 0
            shopping_net = 0
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Miscellaneous Net"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 0
            home = 0
            kids_pets = 0
            misc_net = 1
            misc_pos = 0
            personal_care = 0
            shopping_net = 0
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Miscellaneous Point of Sale(POS)"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 0
            home = 0
            kids_pets = 0
            misc_net = 0
            misc_pos = 1
            personal_care = 0
            shopping_net = 0
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Personal Care"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 0
            home = 0
            kids_pets = 0
            misc_net = 0
            misc_pos = 0
            personal_care = 1
            shopping_net = 0
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Shopping Net"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 0
            home = 0
            kids_pets = 0
            misc_net = 0
            misc_pos = 0
            personal_care = 0
            shopping_net = 1
            shopping_pos = 0
            travel = 0
        elif(trans_reason == "Shopping Point of Sale(POS)"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 0
            home = 0
            kids_pets = 0
            misc_net = 0
            misc_pos = 0
            personal_care = 0
            shopping_net = 0
            shopping_pos = 1
            travel = 0
        elif(trans_reason == "Travel"):
            food_dining = 0
            gas_transport = 0
            grocery_net = 0
            grocery_pos = 0
            health_fitness = 0
            home = 0
            kids_pets = 0
            misc_net = 0
            misc_pos = 0
            personal_care = 0
            shopping_net = 0
            shopping_pos = 0
            travel = 1
        sex = request.form["Gender"]
        if(sex == "0"):
            gender = 0
        elif(sex == "1"):
            gender = 1
        
        prediction = model.predict([[
            credit_card_no,
            trans_amt,
            gender,
            latitude,
            longitude,
            city_pop,
            merch_lat,
            merch_long,
            food_dining,
            gas_transport,
            grocery_net,
            grocery_pos,
            health_fitness,
            home,
            kids_pets,
            misc_net,
            misc_pos,
            personal_care,
            shopping_net,
            shopping_pos,
            travel
        ]])
        output = prediction[0]
        status = ""
        if output == 0:
            status = "factual"
        else:
            status = "fraudulent"
        if status == "factual":
            return render_template('home.html', prediction_text = "Your credit card is {}.".format(status))
        else:
            return render_template('home.html', prediction_text = "Your credit card is {}!".format(status))
```

```python
if __name__ == "__main__":
    app.run(port=8080)
```
